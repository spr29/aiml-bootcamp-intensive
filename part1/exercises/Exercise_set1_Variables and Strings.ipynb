## Exercise 2: String Manipulation

**Question:**

Create a string variable `sentence` with the value "Hello, World!". Then, create another variable that stores the number of characters in `sentence` excluding spaces and print it.

**Example Input:**
```python
sentence = "Hello, World!"
```

**Expected Output:**
```
12
```


````python
# Write your solution here..
````


#### Expand for solution



```python
sentence = "Hello, World!"
char_count = len(sentence.replace(" ", ""))
print(char_count)
```

## Exercise 3: Concatenating Strings

**Question:**

Define two string variables, `first_name` and `last_name`. Concatenate them with a space in between and print the full name.

**Example Input:**
```python
first_name = "John"
last_name = "Doe"
```

**Expected Output:**
```
John Doe
```


````python
# Write your solution here..
````

#### Expand for solution:


```python
first_name = "John"
last_name = "Doe"
full_name = first_name + " " + last_name
print(full_name)
```

## Exercise 5: String Slicing

**Question:**

Define a variable `text` with the value "Python Programming". Extract and print the substring "Programming".

**Example Input:**
```python
text = "Python Programming"
```

**Expected Output:**
```
Programming
```


````python
# Write your solution here..
````

#### Expand for solution

```python

substring = text[7:]  # Starting from index 7 to the end
print(substring)
```

## Exercise 6: Checking Object Types

**Question:**

Create a variable `data` that holds a tuple containing mixed types (an integer, a string, and a float). Print the type of each element in the tuple.

**Example Input:**
```python
data = (1, "apple", 3.14)
```

**Expected Output:**
```
<class 'int'>
<class 'str'>
<class 'float'>
```


````python
# Write your solution here..
````

#### Expand for solution


**Solution:**
```python
data = (1, "apple", 3.14)
for item in data:
    print(type(item))
```

## Exercise 7: Formatting Strings

**Question:**

Define two variables, `age` and `city`, and use an f-string to create a sentence that states "I am {age} years old and live in {city}." Print the sentence.

**Example Input:**
```python
age = 25
city = "New York"
```

**Expected Output:**
```
I am 25 years old and live in New York.
```


````python
# Write your solution here..
````


#### Expand for solution:



```python
age = 25
city = "New York"
sentence = f"I am {age} years old and live in {city}."
print(sentence)
```
