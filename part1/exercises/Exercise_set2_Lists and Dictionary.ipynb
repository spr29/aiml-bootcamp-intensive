## Exercise 1: Unique Elements from Two Lists

**Problem:**
Given two lists of integers, create a new list containing only the unique elements that are present in either of the lists.

**Input:**
```python
list1 = [1, 2, 3, 4, 5]
list2 = [4, 5, 6, 7, 8]
```

**Expected Output:**
```python
unique_elements = [1, 2, 3, 4, 5, 6, 7, 8]
```


````python
# Write your solution here
````


### **Solution:**


````python
list1 = [1, 2, 3, 4, 5]
list2 = [4, 5, 6, 7, 8]
unique_elements = list(set(list1) | set(list2))
unique_elements
````

## Exercise 2: Frequency Count of Elements

**Problem:**
Count the frequency of each element in a list and store the result in a dictionary.

**Input:**
```python
elements = ['apple', 'banana', 'apple', 'orange', 'banana', 'banana']
```

**Expected Output:**
```python
frequency = {'apple': 2, 'banana': 3, 'orange': 1}
```


````python
# Write your solution here..
````

### __Solution:__

````python
elements = ['apple', 'banana', 'apple', 'orange', 'banana', 'banana']
frequency = {}
for item in elements:
    frequency[item] = frequency.get(item, 0) + 1
frequency
````

### Exercise 8: Grouping Elements by Frequency

**Problem:**
Group elements from a list into a dictionary based on their frequency. The keys should be the frequency, and the values should be lists of elements that appear that many times.

**Input:**
```python
elements = ['apple', 'banana', 'apple', 'orange', 'banana', 'banana', 'kiwi']
```

**Expected Output:**
```python
frequency_group = {2: ['apple'], 3: ['banana'], 1: ['orange', 'kiwi']}
```


````python
# Write your solution here
````


### **Solution:**



```python
elements = ['apple', 'banana', 'apple', 'orange', 'banana', 'banana', 'kiwi']
frequency_count = {}
for item in elements:
    frequency_count[item] = frequency_count.get(item, 0) + 1

frequency_group = {}
for item, count in frequency_count.items():
    if count not in frequency_group:
        frequency_group[count] = []
    frequency_group[count].append(item)
```

### Exercise 20: Merge Sorted Lists

**Problem:**
Merge two sorted lists into a single sorted list.

**Input:**
```python
list1 = [1, 3, 5, 7]
list2 = [2, 4, 6, 8]
```

**Expected Output:**
```python
merged_list = [1, 2, 3, 4, 5, 6, 7, 8]
```


````python
# Write your solution here
````


### **Solution:**



```python
list1 = [1, 3, 5, 7]
list2 = [2, 4, 6, 8]
merged_list = []
i, j = 0, 0
while i < len(list1) and j < len(list2):
    if list1[i] < list2[j]:
        merged_list.append(list1[i])
        i += 1
    else:
        merged_list.append(list2[j])
        j += 1
merged_list.extend(list1[i:])
merged_list.extend(list2[j:])
```


```python
duplicates = [2, 3]
```


````python
# Write your solution here
````



### **Solution:**




```python
numbers = [1, 2, 3, 2, 4, 5, 3]
seen = set()
duplicates = set()

    if number in seen:
        duplicates.add(number)
    else:
        seen.add(number)
duplicates = list(duplicates)
```

### Exercise 27: Flattening a Nested Dictionary

**Problem:**
Flatten a nested dictionary, concatenating keys with a separator.

**Input:**
```python
nested_dict = {
    'a': {'b': 1, 'c': 2},
    'd': {'e': 3, 'f': 4}
}
separator = '.'
```

**Expected Output:**
```python
flat_dict = {
    'a.b': 1,
    'a.c': 2,
    'd.e': 3,
    'd.f': 4
}
```


````python
# Write your solution here
````


### **Solution:**



```python
nested_dict = {
    'a': {'b': 1, 'c': 2},
    'd': {'e': 3, 'f': 4}
}
separator = '.'
flat_dict = {}
for key, value in nested_dict.items():
    for sub_key, sub_value in value.items():
        flat_dict[f"{key}{separator}{sub_key}"] = sub_value
```

### Exercise 28: Finding the Most Frequent Element

**Problem:**
Find the most frequent element in a list. If there are ties, return them all in a list.

**Input:**
```python
elements = ['apple', 'banana', 'apple', 'orange', 'banana', 'banana']
```

**Expected Output:**
```python
most_frequent = ['banana']
```


````python
# Write your solution here
````


### **Solution:**



```python
elements = ['apple', 'banana', 'apple', 'orange', 'banana', 'banana']
frequency = {}
for item in elements:
    frequency[item] = frequency.get(item, 0) + 1

max_freq = max(frequency.values())
most_frequent = [key for key, value in frequency.items() if value == max_freq]
```
