### Exercise 1: Simple File Writer
**Description:**

Write a function `write_to_file` that takes a filename and a string. Attempt to write the string to the file, and handle exceptions for file permission errors.

**Input:**
```python
write_to_file("protected_file.txt", "Hello, World!")
```
**Expected Output:**
```
Error: Permission denied when trying to write to the file.
```


````python
# Write your solution here..
````


### **Solution:**


````python
def write_to_file(filename, content):
    try:
        with open(filename, 'r') as file:
            file.write(content)
    except:
        print("Error: Permission denied when trying to write to the file.")

write_to_file("protected_file.txt", "Hello, World!")
````

Make it writable.

````python
def write_to_file(filename, content):
    try:
        with open(filename, 'w') as file:
            file.write(content)
    except:
        print("Error: Permission denied when trying to write to the file.")

write_to_file("protected_file.txt", "Hello, World!")
````

### Exercise 2: Integer Division with Error Handling
**Description:**
Create a function `integer_divide` that takes two integers. Handle the `ZeroDivisionError` and return a message if division by zero is attempted.

**Input:**
```python
integer_divide(10, 0)
```
**Expected Output:**
```
Error: Cannot divide by zero.
```


````python
# Write your solution here..
````


### **Solution:**


````python
def integer_divide(numerator, denominator):
    try:
        return numerator // denominator
    except ZeroDivisionError:
        return "Error: Cannot divide by zero."
        

print(integer_divide(10, 0))
````

### Exercise 3: String to Integer Conversion
**Description:**
Write a function `convert_to_int` that takes a string input and attempts to convert it to an integer. Handle the `ValueError` if the conversion fails, and return an appropriate message.

**Input:**
```python
convert_to_int("abc")
```
**Expected Output:**
```
Error: Invalid input, unable to convert to an integer.
```


````python
# Write your solution here..
````


### **Solution:**


````python
def convert_to_int(input_string):
    try:
        return int(input_string)
    except ValueError:
        return "Error: Invalid input, unable to convert to an integer."

print(convert_to_int("abc"))
````
