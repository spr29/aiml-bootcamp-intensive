## Exercise 1: Chunked File Reader

**Description:**  
Implement a generator that reads a file in chunks of a specified size (in bytes). The generator should yield each chunk as a bytes object.

**Input/Output Example:**
```python
file_gen = read_file_in_chunks('example.txt', 10)
for chunk in file_gen:
    print(chunk)
```
**Expected Output:** (assuming `example.txt` contains the string "Hello World!")
```
Line 1
Line 2
Line 3
```


````python
# Write your solutions here
````


### **Solution:**


````python

def read_file_in_chunks(file_path, chunk_size):
    with open(file_path, 'rb') as file:
        while True:
            chunk = file.read(chunk_size)
            if not chunk:
                break
            yield chunk

file_gen = read_file_in_chunks('example.txt', 8)
for chunk in file_gen:
    print(chunk)

````

### Exercise 2: Custom Range Generator

**Description:**  
Create a custom generator that mimics the behavior of Python’s built-in `range` function. The generator should accept a start, stop, and step parameter and yield numbers in that range.

**Input/Output Example:**
```python
for num in custom_range(1, 10, 2):
    print(num)
```
**Expected Output:**
```
1
3
5
7
9
```


````python
# Write your solution here
````


### **Solution:**


````python

def custom_range(start, stop=None, step=1):
    if stop is None:  # Handle the case where only one argument is provided
        stop = start
        start = 0
    if step == 0:
        raise ValueError("step must not be zero")
    if (start < stop and step < 0) or (start > stop and step > 0):
        return
    while (step > 0 and start < stop) or (step < 0 and start > stop):
        yield start
        start += step

for num in custom_range(1, 10, 2):
    print(num)
````

### Exercise 3: Flatten a Nested List

**Description:**  
Create a generator that flattens a nested list (a list that may contain other lists) into a single list of values. Handle arbitrary levels of nesting.

**Input/Output Example:**
```python
nested_list = [1, [2, [3, 4], 5], 6]
flat_gen = flatten(nested_list)
print(list(flat_gen))
```
**Expected Output:**
```
[1, 2, 3, 4, 5, 6]
```


````python
# Write your solutions here
````


### **Solution:**


````python

def flatten(nested_list):
    for item in nested_list:
        if isinstance(item, list):
            yield from flatten(item)
        else:
            yield item


nested_list = [1, [2, [3, 4], 5], 6]
flat_gen = flatten(nested_list)
print(list(flat_gen))
````
