### Exercise 1: Simple Timing Decorator

**Description:** Create a decorator called `time_it` that measures the time it takes for a function to execute. The decorator should print the execution time.

**Example Input:**
```python
@time_it
def slow_function():
    import time
    time.sleep(2)
    return "Finished"

slow_function()
```

**Expected Output:**
```
Execution time: 2.0 seconds
Finished
```



### **Solution:**


````python

import time

def time_it(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs)
        end_time = time.time()
        print(f"Execution time: {end_time - start_time:.2f} seconds")
        return result
    return wrapper

@time_it
def slow_function():
    time.sleep(2)
    return "Finished"

# Test the function
slow_function()
````

### Exercise 2: Custom Logging Decorator

**Description:** Create a decorator called `log_execution` that logs the function name and its arguments each time the function is called, along with its return value.

**Example Input:**
```python
@log_execution
def add(x, y):
    return x + y

result = add(5, 3)
```

**Expected Output:**
```
Calling add with arguments (5, 3)
add returned 8
```


````python
# Write your solution here
````


### **Solution:**


````python

def log_execution(func):
    def wrapper(*args, **kwargs):
        print(f"Calling {func.__name__} with arguments {args}")
        result = func(*args, **kwargs)
        print(f"{func.__name__} returned {result}")
        return result
    return wrapper

@log_execution
def add(x, y):
    return x + y

# Test the function
result = add(5, 3)
````
