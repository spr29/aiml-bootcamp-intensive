## Exercise: Create a File Operations Package

**Description:**
Create a package named `file_ops` with a module `file_handler.py`. Implement functions to read from and write to text files. Use these functions in the `main.py` to demonstrate file operations.

**Folder Structure:**
```
file_ops/
    ├── __init__.py
    └── file_handler.py
main.py
```

**Input:**
Writing "Hello, World!" to a file named `hello.txt`.

**Expected Output:**
Contents of `hello.txt` printed to the console.

````python
# Write your code here
````


### **Solution:**

Create the following files in a directory named: `file_ops` and creata an empty file: `__init__.py`.



*file_handler.py*
```python
def write_to_file(file_name, content):
    with open(file_name, 'w') as file:
        file.write(content)

def read_from_file(file_name):
    with open(file_name, 'r') as file:
        return file.read()
```

*main.py*
```python
from file_ops.file_handler import write_to_file, read_from_file

write_to_file('hello.txt', 'Hello, World!')
content = read_from_file('hello.txt')
print(content)  # Expected Output: Hello, World!
```
