## Exercise 1: Custom Range Iterator

**Problem Statement:**

Create a custom iterator class called `CustomRange` that mimics the behavior of the built-in `range()` function. The class should accept three parameters: `start`, `stop`, and `step`. It should support iteration over the specified range.

**Example Input:**
```python
for i in CustomRange(1, 10, 2):
    print(i)
```

**Expected Output:**
```
1
3
5
7
9
```


````python
# Write your code here
````


### **Solution:**


````python

class CustomRange:
    def __init__(self, start, stop, step=1):
        self.current = start
        self.stop = stop
        self.step = step

    def __iter__(self):
        return self

    def __next__(self):
        if (self.step > 0 and self.current >= self.stop) or (self.step < 0 and self.current <= self.stop):
            raise StopIteration
        else:
            value = self.current
            self.current += self.step
            return value

````

## Exercise 2: File Line Reader

**Problem Statement:**

Create an iterator class called `FileLineReader` that reads a file line by line. The class should take a file path as an argument and yield each line one at a time when iterated over.

**Example Input:**
Assuming `example.txt` contains:
```
Line 1
Line 2
Line 3
```
```python
reader = FileLineReader('example.txt')
for line in reader:
    print(line.strip())
```

**Expected Output:**
```
Line 1
Line 2
Line 3
```


````python
# Write your code here
````


### **Solution:**


````python
class FileLineReader:
    def __init__(self, filepath):
        self.filepath = filepath
        self.file = open(self.filepath, 'r')

    def __iter__(self):
        return self

    def __next__(self):
        line = self.file.readline()
        if not line:
            self.file.close()
            raise StopIteration
        return line
````
