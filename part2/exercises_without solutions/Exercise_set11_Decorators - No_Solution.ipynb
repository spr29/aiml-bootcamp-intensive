### Exercise 1: Simple Timing Decorator

**Description:** Create a decorator called `time_it` that measures the time it takes for a function to execute. The decorator should print the execution time.

**Example Input:**
```python
@time_it
def slow_function():
    import time
    time.sleep(2)
    return "Finished"

slow_function()
```

**Expected Output:**
```
Execution time: 2.0 seconds
Finished
```
